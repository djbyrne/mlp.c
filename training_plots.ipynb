```python
# Plot training of the C model
import matplotlib.pyplot as plt

# Read the training loss from the file
epochs = []
losses = []

with open('training_loss_c.txt', 'r') as f:
    for line in f:
        epoch, loss, _ = line.strip().split(',')
        epochs.append(int(epoch))
        losses.append(float(loss))

# Plot the training loss
plt.plot(epochs, losses, marker='o')
plt.title('Training Loss over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

def plot_training_loss():
    epochs = []
    losses = []
    cumulative_times = []
    cumulative_time = 0.0

    # Read data from training_loss.txt
    with open('training_loss_c.txt', 'r') as file:
        for line in file:
            if not line.strip():
                continue  # Skip empty lines
            epoch_str, loss_str, duration_str = line.strip().split(',')
            epoch = int(epoch_str)
            loss = float(loss_str)
            duration = float(duration_str)
            cumulative_time += duration

            epochs.append(epoch)
            losses.append(loss)
            cumulative_times.append(cumulative_time)

    # Plot the training loss over cumulative time
    plt.figure(figsize=(10, 6))
    plt.plot(cumulative_times, losses, marker='o', linestyle='-')
    plt.xlabel('Cumulative Time (s)')
    plt.ylabel('Training Loss')
    plt.title('Training Loss Over Time')
    plt.grid(True)
    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    plot_training_loss()
```

```python
import matplotlib.pyplot as plt

# Data for C version
epochs_c = []
losses_c = []
cumulative_times_c = []
cumulative_time_c = 0.0

# Read data from training_loss.txt (C version)
with open('training_loss_c.txt', 'r') as file_c:
    for line in file_c:
        if not line.strip():
            continue  # Skip empty lines
        epoch_str, loss_str, duration_str = line.strip().split(',')
        epoch = int(epoch_str)
        loss = float(loss_str)
        duration = float(duration_str)
        cumulative_time_c += duration

        epochs_c.append(epoch)
        losses_c.append(loss)
        cumulative_times_c.append(cumulative_time_c)

# Data for Python version
epochs_py = []
losses_py = []
cumulative_times_py = []
cumulative_time_py = 0.0

# Read data from training_loss_py.txt (Python version)
with open('training_loss_py.txt', 'r') as file_py:
    for line in file_py:
        if not line.strip():
            continue  # Skip empty lines
        epoch_str, loss_str, duration_str = line.strip().split(',')
        epoch = int(epoch_str)
        loss = float(loss_str)
        duration = float(duration_str)
        cumulative_time_py += duration

        epochs_py.append(epoch)
        losses_py.append(loss)
        cumulative_times_py.append(cumulative_time_py)

# Plot the training loss over cumulative time
plt.figure(figsize=(10, 6))

# Plot for C version
plt.plot(cumulative_times_c, losses_c, marker='o', linestyle='-', label='C Implementation')

# Plot for Python version
plt.plot(cumulative_times_py, losses_py, marker='s', linestyle='--', label='Python Implementation')

plt.xlabel('Cumulative Time (s)')
plt.ylabel('Training Loss')
plt.title('Training Loss Over Time: C vs Python')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
